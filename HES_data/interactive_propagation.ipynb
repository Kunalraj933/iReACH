```python
import pandas as pd
df = pd.read_csv("hospital_communities.csv")
print("SHAPE:", df.shape)
print("\nCOLUMNS:", df.columns.tolist())
print("\nDTYPES:\n", df.dtypes)
print("\nFIRST 5 ROWS:\n", df.head().to_string())
print("\nNULLS:\n", df.isnull().sum())
```

```python
df2 = pd.read_csv("hospital_si_results.csv")
print("SHAPE:", df2.shape)
print("\nCOLUMNS:", df2.columns.tolist())
print("\nDTYPES:\n", df2.dtypes)
print("\nFIRST 5 ROWS:\n", df2.head().to_string())
print("\nNULLS:\n", df2.isnull().sum())
print("\nUNIQUE VALUES (first 3 cols):")
for col in df2.columns[:3]:
    print(f"  {col}: {df2[col].nunique()} unique")
```

```python
agg_hospital_transfers = pd.read_csv("agg_hospital_transfers.csv")
print("SHAPE:", agg_hospital_transfers.shape)
print("\nCOLUMNS:", agg_hospital_transfers.columns.tolist())
print("\nFIRST 5 ROWS:\n", agg_hospital_transfers.head().to_string())
print("\nWEIGHT RANGE:", agg_hospital_transfers["weight"].min(), "—", agg_hospital_transfers["weight"].max())

```

```python
for name, df_check in [
    ("communities", pd.read_csv("hospital_communities.csv")),
    ("si_results",  pd.read_csv("hospital_si_results.csv")),
    ("transfers",   agg_hospital_transfers),
]:
    geo_cols = [c for c in df_check.columns if any(x in c.lower() for x in ["lat","lon","lng","geo","x","y","coord","easting","northing","postcode"])]
    print(f"{name}: {geo_cols if geo_cols else '❌ no geo columns found'}")
```
